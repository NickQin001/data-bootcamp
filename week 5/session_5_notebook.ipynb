```python
comments["word_count"] = comments["comment"].apply(lambda x: len(x.split(" ")))
comments["char_count"] = comments["comment"].apply(lambda x: len(x.replace(" ","")))
comments = pd.concat(
    [comments,pd.get_dummies(comments["category"],drop_first=True)]
    ,axis=1)
```

```python
comments
```

```python
grouped_data = comments.groupby("category")["char_count"].mean()
```

```python
grouped_data
```

```python
comments["mentions_time"] = comments["comment"].apply(lambda x: ("time" in x)*1)
comments
```

```python
comments.columns
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
x = comments.loc[:,["word_count","char_count","mentions_time"]]
y = comments.loc[:,"positive"]
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.coef_, model.intercept_
```

```python
import urllib
from bs4 import BeautifulSoup
```

```python
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:98.0) Gecko/20100101 Firefox/98.0"
}
url = "http://tobypenk.com/scrape-example/"

```

```python
request = urllib.request.Request(url,None,HEADERS)
```

```python
file_req = urllib.request.urlopen(request)
```

```python
file_data = file_req.read()
file_data
```

```python
type(file_data)
```

```python
soup = BeautifulSoup(file_data)
```

```python
veg_recipes = soup.find_all("div",{"class":"veg-recipes"})
```

```python
veg_recipe_stubs = veg_recipes[0].find_all("div",{"class","recipe-stub"})
```

```python
veg_recipe_data = []
for el in veg_recipe_stubs:
  recipe_title = el.find("p",{"class","recipe-title"}).text
  recipe_cals = el.find("p",{"class","recipe-calories"}).text
  recipe_timing = el.find("p",{"class","recipe-timing"}).text
  payload = [recipe_title,recipe_cals,recipe_timing]
  veg_recipe_data.append(payload)

print(veg_recipe_data)
```

```python
veg_recipe_data = pd.DataFrame(veg_recipe_data,columns=["title","cals","timing"],index=None)
```
