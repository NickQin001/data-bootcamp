```python
import random
from sklearn.linear_model import LinearRegression

x1 = list(map(lambda x: x + random.random(),list(range(20))))
x2 = list(map(lambda x: x ** 2 + random.random(),list(range(20))))
y = list(range(23,23+20))
```

```python
import torchvision.transforms as transforms
import torchvision
import torch
```

```python
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding = 4),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])
```

```python
train = torchvision.datasets.CIFAR10(
    root="./data",
    train=True,
    download=True,
    transform=transform
)

trainloader = torch.utils.data.DataLoader(
    train,
    batch_size=100,
    shuffle=True,
    num_workers=2
)
```

```python
from torchvision.transforms import ToTensor
test_transform = transforms.ToTensor()
test = torchvision.datasets.CIFAR10(
    root="./data",
    train=False,
    download=True,
    transform=test_transform
)
testloader = torch.utils.data.DataLoader(
    test,
    batch_size=9,
    shuffle=True
)
```

```python
class_names = ("plane","car","bird","cat","deer","dog","frog","horse","ship","truck")
```

```python
import matplotlib.pyplot as plt
def image_show(img):
  img = img * 0.5 + 0.5
  npimg = img.numpy()
  plt.imshow(np.transpose(npimg, (1, 2, 0)))
  plt.show()
  
```

```python
look = torchvision.datasets.CIFAR10(
    root="./data",
    train=True,
    download=True,
    transform=transform
)
lookloader = torch.utils.data.DataLoader(
    look,
    batch_size=1,
    shuffle=True
)
look_iterator = iter(lookloader)
```

```python
import numpy as np
image, label = next(look_iterator)
image_show(torchvision.utils.make_grid(image))
print(class_names[label.item()])
```

```python
import torch.nn as nn
```

```python
import torch.nn.functional as F
class Net(nn.Module):
  def __init__(self):
    super(Net, self).__init__()
    self.conv1 = nn.Conv2d(3, 6, 5)
    self.pool = nn.MaxPool2d(2, 2)
    self.conv2 = nn.Conv2d(6,16,5)
    self.fc1 = nn.Linear(400, 120)
    self.fc2 = nn.Linear(120, 84)
    self.fc3 = nn.Linear(84,10)

  def forward(self,x):
    x = self.pool(F.relu(self.conv1(x)))
    x = self.pool(F.relu(self.conv2(x)))
    x = x.view(-1,400)
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.fc3(x)
    return x
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
net = Net().to(device)
```

```python
import torch.optim as optim
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
from torchvision.transforms import ToPILImage

def plot_with_predictions(images,labels,predictions,title):
  fig, axes = plt.subplots(3,3, figsize=(8,8))
  fig.subplots_adjust(hspace=0.5,wspace=0.3)
  for i, ax in enumerate(axes.flat):
    ax.imshow(ToPILImage()(images[i]))
    ax.set_title("actual: "+str(class_names[labels[i]]) + " \n pred: " + str(class_names[predictions[i]]))
    ax.axis("off")
  plt.suptitle(title)
  plt.show()
```

```python
EPOCHS = 20
```

```python
for epoch in range(EPOCHS):
  running_loss = 0.0
  for i, data in enumerate(trainloader, 0):
    inputs, labels = data
    inputs, labels = inputs.to(device), labels.to(device)

    optimizer.zero_grad()

    outputs = net(inputs)
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()

    running_loss += loss.item()

  test_iterator = iter(testloader)
  images, labels = next(test_iterator)
  images, labels = images.to(device), labels.to(device)

  with torch.no_grad():
    outputs = net(images)
    _, predicted = torch.max(outputs.data, 1)
    predicted = predicted.cpu().numpy()

  images, labels = images.cpu(), labels.cpu().numpy()
  plot_with_predictions(images, labels, predicted, "epoch "+str(epoch+1))

```

```python
from PIL import Image
import requests
from io import BytesIO
import torchvision.transforms as transforms

url = "https://d1jyxxz9imt9yb.cloudfront.net/medialib/3078/image/s1300x1300/IP202207_GlassFrogs_009_365211_reduced.jpg"
response = requests.get(url)
img = Image.open(BytesIO(response.content)).convert("RGB")

preprocess = transforms.Compose([
    transforms.Resize((32,32)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])

img_data = preprocess(img).unsqueeze(0).to(device)

with torch.no_grad():
  output = net(img_data)
  _, prediction = torch.max(output.data,1)
  prediction = prediction.cpu().numpy()[0]

plt.imshow(img)
plt.title("predicted "+class_names[prediction])
plt.axis("off")
plt.show()



```
