```python
def doubler(n):
    return n*2

test_doubler = {
    4:8,
    8:16,
    -5:-10,
    1:2,
    0:0
}

for i in test_doubler:
    if  doubler(i) != test_doubler[i]:
        print("error with pair " +str(i)+ " " +str( test_doubler[i]))
```

```python
# filter
x = list(range(100))

def custom(n):
    if n% 17 ==0 or n %10 ==3: return True
    return False

evens= list(filter(lambda i: i%17 ==0,x))
evens

el2=list(filter(custom,x))
el2
```
