```python
# Question 1


num1=float(input("Please enter the first number: "))
num2=float(input("Please enter the second number: "))
operator=str(input("*Vaild operators are: add, subtract, multiply, divide, power, modulo.* Please enter the operatoer: ")).lower()

def  operator_function( num1,  num2, operator):
  if  operator=="add":
    return num1+num2
  elif operator=="subtract":
    return num1-num2
  elif operator=="multiply":
    return num1*num2
  elif operator=="divide":
    return num1/num2
  elif operator=="power":
    return num1**num2
  elif operator=="modulo":
    return num1%num2
  else:
    return 'ERROR! Please enter two vaild numbers and one valid operator! Vaild operators are: add, subtract, multiply, divide, power, modulo'


print(operator_function( num1,  num2, operator))
```

```python
# Question 2 discount calculator


price=float(input("Please enter the amount of price: "))

def discount_calculator(price):
  if price>=0:
    if price>50:
      total=price*0.90
      return round(total,2)
    else:
      total=price
      return round(total,2)
  else:
    return 'unable to determine. Please enter a vaild price!'

print(f"The total price is {discount_calculator(price)}")
```

```python
#Question 3 currency converter
import requests

from_cur = str(input("Please enter the currency converting from: ")).upper()

to_cur = str(input("Please enter the currency converting to: ")).upper()

amount_of_money = float(input("Please eneter the amount of money: "))

response = requests.get( f"https://api.frankfurter.app/latest?amount={amount_of_money}&from={from_cur}&to={to_cur}")
# print(response.status_code)

# returns json object for the specific rate in the currency rate list
print( f"{amount_of_money} {from_cur} is {response.json()['rates'][to_cur]} {to_cur}")
```

```python
#Question 4 loan eligibility calculator

# A borrower’s credit score is a numerical representation of their creditworthiness
credit_sourcing=str(input("*Valid options: high, low* Please enter the borrower’s credit score : ")).lower()

# Lenders need to verify the borrower’s income in order to assess their ability to repay the loan
income= str(input("*Valid options: high, low* Please enter the borrower’s income level: ")).lower()

def loan_qligibility_calculator(credit_sourcing, income):
  if credit_sourcing =="high":
    if income=="high":
      return True
    elif income=="low":
      return False
    else:
      return "Please enter valid option"
  elif credit_sourcing =="low":
    if income=="high":
      return False
    elif income=="low":
      return False
    else:
      return "Please enter valid option"

print(loan_qligibility_calculator(credit_sourcing, income))


```

```python
# challenge
'''
question : If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.
Find the sum of all the multiples of 3 or 5 below 1000.
'''

total_sum = 0

for i in range(1000):
    if (i%3 == 0 or i%5 == 0):
        total_sum = total_sum+i
        
print (total_sum)

```
