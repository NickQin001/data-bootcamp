```python
from bs4 import BeautifulSoup
import re
import pandas as pd

#from os import listdir
#from os.path import isfile, join
import urllib

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.expected_conditions import visibility_of_element_located

```

```python
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent,}

url = "https://www.walmart.com/search?q=lego"

request = urllib.request.Request(url,None,headers)
f = urllib.request.urlopen(request)
file_data = f.read()

soup = BeautifulSoup(file_data)
print(len(soup.find_all("div",{"data-item-id":True})))
soup
```

```python
def fetch_page_data_wm(url,force_local = False):

    browser = webdriver.Chrome()
    browser.get(url)
    browser.execute_script("window.scrollTo(0, 100)")
    request = (
        WebDriverWait(driver=browser, timeout=10)
        .until(visibility_of_element_located((By.CSS_SELECTOR, "div")))
        .text
    )
    file_data = browser.page_source
    browser.close()
        
    return file_data
```

```python
urls = ["https://www.walmart.com/search?q=lego"]
for i in range(2,10):
    urls.append("https://www.walmart.com/search?q=lego&page="+str(i))
urls
```

```python
for url in urls:
    print(url)
    page_data = fetch_page_data_wm(url)
    soup = BeautifulSoup(page_data)
    print(len(soup.find_all("div",{"data-item-id":True})))
```
